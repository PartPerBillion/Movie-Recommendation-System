```python
import pandas as pd

m = pd.read_csv('movies.csv')
m.head()
```

```python
m.columns
```

```python
n = m[['genres', 'keywords', 'tagline', 'cast', 'director']]
f = ['genres', 'keywords', 'tagline', 'cast', 'director']

for i in f:
    n[i] = n[i].fillna(' ')
    
n.head()
```

```python
cf = n['genres']+' '+n['keywords']+' '+n['tagline']+' '+n['cast']+' '+n['director']
cf
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

t = TfidfVectorizer()
v = t.fit_transform(cf)
print(v)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cs = cosine_similarity(v)
print(cs)
```

```python
cs.shape
```

```python
mn = m['title'].tolist()
mn
```

```python
p = input('type the movie you watched from above :')
```

```python
import difflib

cm = difflib.get_close_matches(p, mn)
c = cm[1]
```

```python
mi =m[m.title == c]['index'].values[0]
print(mi)
```

```python
ss = list(enumerate(cs[mi]))
# ss
```

```python
len(ss)
```

```python
ssm = sorted(ss, key = lambda x:x[1], reverse =True)
print(ssm)
```

```python
print('Movies Suggested for you : \n')

i = 1

for mo in ssm:
    ind = mo[0]
    tfi = m[m.index == ind]['title'].values[0]
    if (i<30):
        print(i,' ',tfi)
        i+=1
```
